```python
import plotly.express as px

# Crear un boxplot interactivo para las columnas numéricas
fig = px.box(df,
             y=df.select_dtypes(include=['float64', 'int64']).columns,
             title="Boxplot Interactivo de Columnas Numéricas",
             labels={"value": "Valor", "variable": "Características"})  # Etiquetas para los ejes

# Mostrar el gráfico
fig.show()
```

```python
# Crear un gráfico de dispersión para comparar dos variables numéricas
fig = px.scatter(df, x='age', y='bmi', color='smoker',  # Aquí puedes cambiar 'age' y 'bmi' por tus columnas de interés
                 title="Gráfico de Dispersión: Edad vs. IMC (BMI) con Outliers",
                 labels={"age": "Edad", "bmi": "Índice de Masa Corporal (BMI)"})  # Etiquetas para los ejes

# Mostrar el gráfico
fig.show()
```
